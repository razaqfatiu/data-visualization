```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('https://tinyurl.com/ChrisCoDV/world_population.csv', index_col=0)

data.head()
# data.describe()
# data.isna().sum()
```

We are going to transpose the dataframe because we want the countries as the rows.

Then we'll have to remove 'Country Code', 'Indicator Name', 'Indicator Code' becasue these are now going to be rows with descriptive data.

```python
# print(data.columns)

df = data.transpose()

# print(data.columns)
# data.index

selected_drop = ['Country Code', 'Indicator Name', 'Indicator Code']

df = df.drop(selected_drop, axis=0)

df.head(10)
# df.describe()
```

```python
df = df.reindex(df.mean().sort_values(ascending=False).index, axis=1)
plt.figure(figsize=(10,10))
x_pos = np.arange(len(df.columns))

plt.bar(x_pos, df.mean(), align='center')
plt.xticks(x_pos, df.columns)
plt.xlabel('Years')
plt.ylabel('Countries')
```

```python
df.mean()

categories = ['Extreme', 'High', 'Moderate', 'Low']
selected_categories = [[] for i in categories]

for country in df.columns:
  avg_size = df[country].mean()

  if avg_size > 10000:
    category = 0
  elif avg_size > 1000:
    category = 1
  elif avg_size > 500:
    category = 2
  else:
    category = 3
  selected_categories[category].append(country)

print(selected_categories)

for i, selected in enumerate(selected_categories):
  plt.figure(figsize=(10,10))
  x_pos = np.arange(len(df[selected].columns))
  plt.bar(x_pos, df[selected].mean(), align='center')
  plt.xticks(x_pos, df[selected].columns)
  plt.xlabel('Countries')
  plt.ylabel('Density')
```

```python
selected = ['Aruba',	'Andorra',	'Afghanistan',	'Angola',	'Albania'	]

plt.figure(figsize=(10,10))
x_pos = np.arange(len(df[selected].columns))
plt.bar(x_pos, df[selected].mean(), align='center')
plt.xticks(x_pos, df[selected].columns)
plt.xlabel('Countries')
plt.ylabel('Density')
```

```python
selected = ['Aruba',	'Andorra',	'Afghanistan',	'Angola',	'Albania'	]
years = ['1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970']

dtf=df.loc[years]

plt.figure(figsize=(10,10))
x_pos = np.arange(len(dtf[selected].columns))
plt.bar(x_pos, dtf[selected].mean(), align='center')
plt.xticks(x_pos, dtf[selected].columns)
plt.xlabel('Countries')
plt.ylabel('Density')
plt.title('Mean population density from 1961-1970')
```
