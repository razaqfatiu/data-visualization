 Create a single visualisation showing scatter subplots for all pairs of measurements. You may find the following line of code useful:
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width'] Comment on the strongest correlation, visually.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.read_csv('https://tinyurl.com/ChrisCoDV/iris.csv')

selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

pd.plotting.register_matplotlib_converters()

counter = 1
fig = plt.figure(figsize=(12, 12))
fig.suptitle('Flower size correlation', fontsize=14, position=(0.5, 1.0))
for i, name_i in enumerate(selected):
    for j in range(i + 1, len(selected)):
        name_j = selected[j]
        sub = fig.add_subplot(4, 4, counter)
        sub.set_title(name_i + ' vs ' + name_j, fontsize=10)
        sub.scatter(data[name_i], data[name_j], s=0.5)
        counter += 1
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


data = pd.read_csv('https://tinyurl.com/ChrisCoDV/iris.csv')

pd.plotting.register_matplotlib_converters()
print(data.head())

selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

plt.figure(figsize=(10, 10))
corr = data[selected].corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(220, 20, n=200), square=True, annot=True,
                 annot_kws={"size": 8})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```
