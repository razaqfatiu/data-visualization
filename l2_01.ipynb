```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('https://tinyurl.com/ChrisCoDV/Pages/DailyHits.csv', index_col=0)

data.head()
data.describe()
```

The time period under investigation is 365 days (everyday of the year 2019).

The data consists of 167 pages for everyday of the year.

Page 001 has the maximum hits while many pages like 006,007,..., have lowest minimum hits.


```python
data = data.reindex(data.sum().sort_values(ascending=False).index, axis=1)

selected = []

for x in range(1, 16) :
  selected.append(str(x).zfill(3))

data[selected]


# print(data.sum().head(15))

x_pos = np.arange(len(data.columns))
plt.bar(x_pos, data.sum(), align='center')
plt.xticks(x_pos, data.columns)
plt.xlabel('Pages')
plt.ylabel('Number of Views')
plt.show()

# x_pos = np.arange(len(data[selected].columns))
# plt.bar(x_pos, data[selected].sum(), align='center')
# plt.xticks(x_pos, data[selected].columns)
```

```python
# data = pd.read_csv('https://tinyurl.com/ChrisCoDV/Pages/DailyHits.csv', index_col=0)

categories = ['High', 'Medium', 'Low']
cat_selection = [[] for c in range(len(categories))]

# print(cat_selection)

for cat in data.columns:
  total_col_sum = data[cat].sum()

  if total_col_sum > 40000:
    category = 0
  elif total_col_sum > 10000:
    category = 1
  else:
    category = 2
  cat_selection[category].append(cat)

# print(cat_selection)


# data.head()

# data[data.columns].sum().sample(10)

x_pos = np.arange(len(cat_selection[0]))
plt.bar(x_pos, data[cat_selection[0]].sum(), align='center')
plt.xticks(x_pos, cat_selection[0])
plt.xlabel('Pages')
plt.ylabel('Views')
plt.title('High category Page per view')
plt.show()
```

```python
data = pd.read_csv('https://tinyurl.com/ChrisCoDV/Pages/DailyHits.csv', index_col=0)

data = data.reindex(data.sum().sort_values(ascending=False).index, axis=1)

explodeList = []

selected = []

columns = data.columns

data['Low_Hits'] = [0] * len(data.index)

for page in columns:
  total_hits = data[page].sum()
  if total_hits > 10000:
    selected.append(page)
    explodeList.append(0)
  else:
    data['Low_Hits'] += data[page]

selected.append('Low_Hits')
explodeList.append(0.1)

print(explodeList)

plt.figure(figsize=(10,10))
plt.pie(data[selected].sum(), labels=selected, autopct='%1.1f%%', startangle=90, explode=explodeList)
plt.title('Total Page and hits', fontsize=20)
plt.legend(loc=2)
plt.show()
```
